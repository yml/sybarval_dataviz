# Sybarval émission carto


```python
import geopandas as gpd

communes = gpd.read_file("_data_cache/communes-nouvelle-aquitaine.geojson")
communes[communes["code"]=="33005"]
```

```python
COBAN=[
    "33284", #'MIOS',
    "33555", #'MARCHEPRIME',
    "33051",# 'BIGANOS',
    "33019", # 'AUDENGE',
    "33229", # 'LANTON',
    "33005", # 'ANDERNOS-LES-BAINS',
    "33011", # 'ARES',
    "33236", #'LEGE-CAP-FERRET',
    ]

VAL_DE_L_EYRE = [
    "33042", # 'BELIN-BELIET',
    "33029", # 'LE BARP',
    "33260", # 'LUGOS',
    "33436", # 'SAINT-MAGNE',
    "33498", # 'SALLES',
    ]

COBAS = [
    "33009", # 'ARCACHON',
    "33199", # 'GUJAN-MESTRAS',
    "33529", # 'LA TESTE-DE-BUCH',
    "33527", # 'LE TEICH'
    ]
    
SYBARVAL = COBAN + COBAS + VAL_DE_L_EYRE
```

```python
geo_sybarval = communes[communes["code"].isin(SYBARVAL)]
geo_sybarval
```

```python
ax = geo_sybarval.plot()
ax.set_axis_off()
```

```python
import numpy as np
import pandas as pd

URL = "https://data.ademe.fr/data-fair/api/v1/datasets/igt-pouvoir-de-rechauffement-global/full"
FILE  = "_data_cache/igt_ademe_2016.csv"

# df = pd.read_csv(URL)
igt_df = pd.read_csv(FILE)
```

```python
igt_sybarval = igt_df[igt_df["INSEE commune"].isin(SYBARVAL)]
igt_sybarval
```

```python
joined_sybarval = geo_sybarval.set_index("code").join(igt_sybarval.set_index("INSEE commune"))
```

```python
ax = joined_sybarval.plot(
    'Routier', legend=True,
    legend_kwds={'label': "tCO2/an", 'orientation': "vertical"},
    cmap='OrRd',
)
ax.set_axis_off()
ax.set_title("Les émissions de GES pour le transport routier")
```

```python
all_emissions = [
    "Autres transports",
    "Routier",
    "Autres transports international",
    "Agriculture",
    "CO2 biomasse hors-total",
    "Déchets",
    "Energie",
    "Industrie hors-énergie",
    "Résidentiel",
    "Tertiaire",
]
```

```python
ax = joined_sybarval.plot(
    joined_sybarval[all_emissions].sum(axis=1),
    legend=True,
    legend_kwds={'label': "tCO2/an", 'orientation': "vertical"},
    cmap='OrRd',
)
ax.set_axis_off()
ax.set_title("GES  toutes les émissions")
```
