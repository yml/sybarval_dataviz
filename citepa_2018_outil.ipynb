##Inventaires spatialisés

### Format

Ce format de rapportage consiste à spatialiser (c’est-à-dire cartographier dans un cadastre, sur une carte de communes, etc.) des émissions déjà estimées dans l’inventaire. Il s’agit alors d’une autre représentation des données d’émissions, où les émissions sont replacées à leur lieu d’origine.

Nous réalisons aussi des projets de spatialisation d’émissions pour des territoires (pays, régions…): découvrez notre offre de prestations sur la spatialisation.

### Contexte

#### Spatialisation des émissions de polluants
 

Dans le cadre de la CEE-NU, les émissions de polluants doivent être régulièrement spatialisées au sein de la Grille Emep. 
Au niveau des régions, les AASQA peuvent réaliser des cadastres d’émission: consulter leurs sites respectifs.

##### Spatialisation des émissions de gaz à effet de serre
 

Dans le cadre de la CCNUCC, aucune spatialisation des émissions de GES n’est requise. 
Dans le cadre du décret n°2016-849 du 28 juin 2016, complété par l’arrêté du 4 août 2016, les intercommunalités à fiscalité propre de plus de 20 000 habitants doivent inclure dans leur Plan Climat Air Energie Territorial (PCAET) un diagnostic relatif à l’estimation des émissions territoriales de gaz à effet de serre et de polluants atmosphériques. Le Citepa a réalisé la spatialisation, par commune, des émissions de GES de 2016 et 2018 afin de fournir à ces intercommunalités, si besoin, une estimation par défaut.
Le Citepa a aussi participé à la réalisation et à la mise à jour de l’Inventaire National Spatialisé (INS).

### Substances couvertes

Gaz à effet de serre : CO2, CH4, N2O, HFC, PFC, SF6, NF3, total en CO2e.

Polluants : le SO2, les NOX, le NH3, les COVNM, le CO, les métaux lourds (As, Cd, Cr, Cu, Hg, Ni, Pb, Se, Zn), les poussières totales en suspension (TSP), les particules fines (PM10 et PM2,5) , le black carbon (carbone suie, BC) et certains composés organiques persistants (POP) tels que les dioxines et les furannes (PCDD/F), les hydrocarbures aromatiques polycycliques (HAP) sous forme spéciée (BaP, BbF, BkF et IndPy), les polychlorobiphényles (PCB) et l’hexachlorobenzène (HCB).

```python
import pandas as pd

FILE = "_data_cache/SYNTHESE_resultats_emissions_2018_IGT_d.xlsx"

df_prg = pd.read_excel(FILE, sheet_name="PRG")
df_population = pd.read_excel(FILE, sheet_name="Population")
df_communes = pd.read_excel(FILE, sheet_name="Communes")
```

```python
df = df_communes.set_index("CODE_COM_2020",drop=False).join(df_prg.set_index("CODE_COM_2020").drop("EPCI", axis=1)).join(df_population.set_index("CODE_COM_2020"))
```

```python

COBAN=[
    "33284", #'MIOS',
    "33555", #'MARCHEPRIME',
    "33051",# 'BIGANOS',
    "33019", # 'AUDENGE',
    "33229", # 'LANTON',
    "33005", # 'ANDERNOS-LES-BAINS',
    "33011", # 'ARES',
    "33236", #'LEGE-CAP-FERRET',
    ]

VAL_DE_L_EYRE = [
    "33042", # 'BELIN-BELIET',
    "33029", # 'LE BARP',
    "33260", # 'LUGOS',
    "33436", # 'SAINT-MAGNE',
    "33498", # 'SALLES',
    ]

COBAS = [
    "33009", # 'ARCACHON',
    "33199", # 'GUJAN-MESTRAS',
    "33529", # 'LA TESTE-DE-BUCH',
    "33527", # 'LE TEICH'
    ]
    
SYBARVAL = COBAN + COBAS + VAL_DE_L_EYRE
```

```python
df_sybarval = df[df["CODE_COM_2020"].isin(SYBARVAL)]
df_sybarval
```

```python
[col for col in df_sybarval.columns]
```

```python
all_cols =  [
    'Energie',
    'Industrie_hors_energie',
    'Residentiel',
    'Tertiaire',
    'Dechets',
    'Routier',
    'Autres_Transports',
    'Autres_Transports_international',
    'Agriculture',
]

transports = [
    'Routier',
    'Autres_Transports',
    'Autres_Transports_international',
]
```

```python
emissions_transports = df_sybarval[transports].agg("sum").sum()
print(f"{emissions_transports=}")
```

```python
emissions_toutes  = df_sybarval[all_cols].agg("sum").sum()
print(f"{emissions_toutes=}")
```

```python
transport_percent = emissions_transports/emissions_toutes
print(f"{transport_percent=}")
```
