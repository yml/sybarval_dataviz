# Inventaire de gaz à effet de serre territorialisé

## Description

Le Ministère en charge de l’Environnement a chargé le Centre Interprofessionnel Technique d’Etudes de la Pollution Atmosphérique (CITEPA) d’une mission de « territorialisation » – ou « spatialisation » - de l’inventaire national de GES. La résolution spatiale est communale, structures stables dans le temps et ensuite agrégeables par EPCI. Il est établi à partir à la fois à partir d’une décomposition des émissions nationales de GES au niveau communal et d’informations déjà spatialisées.

Cet inventaire spatialisé a vocation à donner des ordres de grandeur pertinents pour l’année 2016. Il n’a pas le niveau de finesse et de précision des travaux d’inventaire réalisés au niveau territorial par les associations agréées de surveillance de la qualité de l’air (www.atmo-france.org), les agences régionales de l’énergie (www.rare.fr), ou des travaux nationaux à hautes résolutions spatiales et temporelles (projet de grande ampleur type Inventaire National Spatialisé - INS).

Il peut constituer un inventaire « par défaut » pour les collectivités qui n’auraient pas accès à des inventaires GES territoriaux déjà bien établis. Les données spatialisées concernant les autres polluants de l’air (particules, oxydes d’azote, …) sont réglementairement et périodiquement calculées et mises à disposition par les associations agréées de surveillance de la qualité de l’air. Elles peuvent être intégrées dans le diagnostic PCAET des collectivités.

Les observatoires régionaux de l'énergie mettent à disposition des territoires des données liées à la consommation et à la production d'énergie via leurs plateformes régionales, accessibles depuis www.rare.fr.</p>

Le jeu de données proposé ici intègre les effets de l'ensemble des gaz à effet de serre, en détaillant les émissions par commune et par secteur. Si les données sont diffusées à la commune, elles ne sont représentatives qu'à une échelle agrégées à l'EPCI.

## Download and load the data

```python
import pandas as pd

URL = "https://www.data.gouv.fr/fr/datasets/r/42279418-4e1f-4c8d-bea4-919af015ae20"
FILE = "igt_2016_global.csv"

# data = pd.read_csv(URL)
## Use  the local CSV instead of the one from data.gouv.fr
## This is used as a local cache
data =pd.read_csv(FILE)
data["INSEE commune"] = data["INSEE commune"].astype("str")
```

## Definition du territoire

```python
COBAN=[
    'MIOS',
    'MARCHEPRIME',
    'BIGANOS',
    'AUDENGE',
    'LANTON',
    'ANDERNOS-LES-BAINS',
    'ARES',
    'LEGE-CAP-FERRET',]

VAL_DE_L_EYRE = [
    'BELIN-BELIET',
    'LE BARP',
    'LUGOS',
    'SAINT-MAGNE',
    'SALLES',]

COBAS = [
    'ARCACHON',
    'GUJAN-MESTRAS',
    'LA TESTE-DE-BUCH',
    'LE TEICH']

SYBARVAL = COBAN + COBAS + VAL_DE_L_EYRE
```

```python
syberval_df = data[data["Commune"].isin(SYBARVAL) & data["INSEE commune"].str.startswith("33")]
syberval_df
```

```python
import seaborn as sns
sns.set_theme()

sns.displot(data=syberval_df)
```

```python
import matplotlib.pyplot as plt

# declaring exploding pie
explode = [0.3 for n in syberval_df["Commune"]]
plt.pie(syberval_df["Routier"], labels=syberval_df["Commune"],autopct='%.0f%%', explode=explode)
plt.show()
```

```python
transport_cols = [
    "Autres transports",
    "Routier",
    "Autres transports international",]

syberval_df[transport_cols].fillna(0).agg("sum")
```

```python
sns.displot(data=syberval_df[transport_cols].fillna(0))
```

```python
transport = syberval_df[transport_cols].fillna(0).agg("sum").sum()
print(f"{transport=}")
```

```python
all_cols = [
    "Autres transports",
    "Routier",
    "Autres transports international",
    "Agriculture",
    "CO2 biomasse hors-total",
    "Déchets",
    "Energie",
    "Industrie hors-énergie",
    "Résidentiel",
    "Tertiaire",
]
syberval_df[all_cols].fillna(0).agg("sum")
```

```python
sns.histplot(data=syberval_df[all_cols].fillna(0))
```

```python
syberval_df.set_index(["Commune"])
```

```python
import matplotlib.pyplot as plt

# declaring exploding pie
explode = [0.4 for n in syberval_df["Commune"]]
plt.pie(syberval_df[all_cols].sum(axis=1), labels=syberval_df["Commune"],autopct='%.0f%%', explode=explode)
plt.show()
```

```python
all = syberval_df[all_cols].fillna(0).agg("sum").sum()
print(f"{all=}")
```

```python
transport_percent = transport/all
print(f"{transport_percent=}")
```
